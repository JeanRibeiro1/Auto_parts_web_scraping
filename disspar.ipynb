Importações necessárias 

```python
from requests_html import HTMLSession
import csv
import pandas as pd
```

Loop nas paginas para pegar todos os links dos produtos

```python
#iniciado a lista de produtos
products = []

#iniciando a variavel de pagina para começar já na primeira
page = 1

while True:
    #URL da pagina que vai mudando o numero da pagina conforme vai rolando o loop
    url = f'https://www.disppar.com.br/loja/catalogo.php?loja=984694&categoria=33&pg={page}'
    
    #codigo para iniciar a sessão, seria como abrir um navegador
    s = HTMLSession()
    r = s.get(url)
    
    #xpath da area que contem todos os produtos de cada pagina
    xpath = '//*[@class="showcase-catalog clearfix col-sm-9"]'
    #objetos dos conteiner
    productss = r.html.xpath(xpath, first=True)
    #junta os objetos do container em uma lista
    products.append(productss)
    # Checa se tem mais paginas e se tiver ele da break
    if "Próxima" not in r.html.html:
        break
    #adiciona um 1 a pagina para a iteração na URL
    page += 1
print(page)

```

Iteração nos produtos para pegar o Nome e Preço dos produtos

```python
#inicia a lista que vai colocar o nome e o preço
data = []
#loop na lista de produtos que cada uma contem a lista de objetos de cada pagina
for product in products:
    #for dentro de cada objeto para pegar os links
    for link in product.absolute_links:
        #pega os links e abre a sessão
        r = s.get(link)
        
        #Testa se há preço no link do produto e se não tiver é para aí e iniciado outro loop
        try:
            price = r.html.xpath('//*[@id="product-container"]/span[2]/meta[1]', first=True).attrs['content']
        except:
            continue
            
        name = r.html.find('h1.product-name',first=True).text
        marca = r.html.find('span.dados-valor.brand', first=True).text
        print(name)
        print(price)
        print(marca)
        
        data.append([name, price, marca])
#criando dataframe
df = pd.DataFrame(data, columns = ['Nome do Produto', 'Preço', 'Marca'])

#colocando os dados do dataframe em um arquivo CSV
df.to_csv("dados_site.csv", sep=';',index=False, encoding = 'utf-8-sig')        
```
